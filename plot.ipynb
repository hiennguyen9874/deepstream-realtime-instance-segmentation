```python
import matplotlib.pyplot as plt
```

```python
data = [
    ("YOLOv5", "YOLOv5n", 247.97, 0.27, 0.234, "r"),
    ("YOLOv5", "YOLOv5s", 183.77, 0.372, 0.317, "r"),
    ("YOLOv5", "YOLOv5m", 119.58, 0.445, 0.373, "r"),
    ("YOLOv5", "YOLOv5l", 85.80, 0.484, 0.401, "r"),
    ("YOLOv5", "YOLOv5x", 54.04, 0.502, 0.415, "r"),
    ("YOLOv7", "YOLOv7-Seg", 84.12, 0.49, 0.406, "g"),
    ("YOLOv7", "YOLOv7-Seg-Dev", 84.59, 0.505, 0.414, "g"),
    ("YOLOv7", "YOLOv7x-Seg", 60.29, 0.511, 0.421, "g"),
]
```

```python
data_line = (
    ("YOLOv5n", "YOLOv5s", 'r', 'dashdot'),
    ("YOLOv5s", "YOLOv5m", 'r', 'dashdot'),
    ("YOLOv5m", "YOLOv5l", 'r', 'dashdot'),
    ("YOLOv5l", "YOLOv5x", 'r', 'dashdot'),
    ("YOLOv7-Seg", "YOLOv7x-Seg", 'g', "dashed"),
)
```

```python
group_data = {x[1]: x for x in data}
```

```python
import itertools

marker = itertools.cycle((",", "+", ".", "o", "*"))

for group_name, model_name, fps, map_box, map_mask, color in data:
    # plt.scatter(fps, map_box, label=model_name, color=color, marker=next(marker))
    plt.scatter(fps, map_mask, label=model_name, color=color, marker=next(marker))

for from_name, to_name, color, linestyle in data_line:
    plt.plot(
        (group_data[from_name][2], group_data[to_name][2]),
        # (group_data[from_name][3], group_data[to_name][3]),
        (group_data[from_name][4], group_data[to_name][4]),
        color=color,
        linestyle=linestyle,
    )

plt.legend()
plt.xlabel("Deepstream FPS")
plt.ylabel("mAP 0.5:0.95")
plt.show()
```
